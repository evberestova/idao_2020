```python
count = Counter()
sat_id_to_mode = dict()
for sat_id in set(train_df.sat_id):
    sat_df = train_df[train_df.sat_id == sat_id]
    xs = sat_df.x.values
    ids = np.argsort(xs)
    ids_diff = np.abs(ids[1:] - ids[:-1])
    mode = sps.mode(ids_diff).mode[0]
    count[mode] += 1
    sat_id_to_mode[sat_id] = mode
```
